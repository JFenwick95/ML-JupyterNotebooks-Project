```python
# install...
# wheel
# pandas
# pyarrow
# ipywidgets
# matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv("ds_salaries.csv")
```

```python
num_bins = 20
plt.figure(figsize=(8,4))
plt.hist(dataset['salary_in_usd'], bins=num_bins, color='blue', edgecolor='black', alpha=0.7)
plt.xlabel('Salary in USD')
plt.ylabel('Number of Employees')
plt.title('Data Scientist Salary Ranges')
plt.show()
```

```python
lvl_of_experience = {'EN': 1, 'MI': 2, 'SE': 3, 'EX': 4}
dataset['Level of Experience'] = dataset['experience_level'].map(lvl_of_experience)
plt.figure(figsize=(8, 8))
plt.scatter(dataset['Level of Experience'], dataset['salary_in_usd'], 
            alpha=0.5, color='blue', edgecolors='black', s=1)
plt.xlabel('Experience Level')
plt.ylabel('Salary')
plt.title('Experience vs. Salary')
plt.xticks([1, 2, 3, 4], ['EN', 'MI', 'SE', 'EX'])
least_squares_polynomial_fit = np.polyfit(dataset['Level of Experience'], dataset['salary_in_usd'], 1)
regression_line = np.polyval(least_squares_polynomial_fit, [1, 2, 3, 4])
plt.plot([1, 2, 3, 4], regression_line, 'r', linewidth=2)
plt.show()
```
